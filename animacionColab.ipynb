```
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import animation, rc
from IPython.display import HTML


def plots():
  fig, ax = plt.subplots()
  plt.close()
  ax.set_xlim(( 0, 50))
  ax.set_ylim((-2, 2))
  line, = ax.plot([], [], lw=2)
  return(fig,line)

fig,line=plots()


# initialization function: plot the background of each frame
def init():
    line.set_data([], [])
    return (line,)

def wave(i):
    x = np.linspace(0, 50, 1000)
    k=2*np.pi/10.0
    w=2*np.pi/20.0
    y = np.sin(k* (x)-w*i)
    return (x,y)


# animation function. This is called sequentially  
def animate(i):
    x,y = wave(i)
    line.set_data(x, y)
    return (line,)
  

anim = animation.FuncAnimation(fig, animate, init_func=init,
                             frames=100, interval=50, blit=True)

# Note: below is the part which makes it work on Colab
rc('animation', html='jshtml')
anim
```

```
from IPython.display import HTML
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

#data = np.random.rand(100,50,50)
#-----------------
from numpy import exp,arange
from pylab import meshgrid,cm,imshow,contour,clabel,colorbar,axis,title,show
nx=100
# the function that I'm going to plot

              
def z_func(i,x,y):
  r=np.sqrt((x-20)**2+y**2)
  r2=np.sqrt((x-80)**2+y**2)
  k=2*np.pi/10.0
  w=2*np.pi/20.0
  gauss=np.exp(-0.5*(r/35.0)**2)
  gauss2=np.exp(-0.5*((r2)/35.0)**2)
  seno=np.sin(k*r-w*i)*gauss+np.sin(k*(r2)-w*i)*gauss2
  return (seno)            
              
 #g=1#np.exp(-0.5)
    
 
x = arange(0,nx,1)
y = arange(0,nx,1)
X,Y = meshgrid(x, y) # grid of point
#Z = z_func(X, Y) # evaluation of the function on the grid
#-----------------

fig, ax = plt.subplots(figsize=(5,5))


ax.set_xlim((0, nx))
ax.set_ylim((0, nx))

im = ax.imshow(z_func(0,X,Y),interpolation="spline16")

def init():
    im.set_data(z_func(0,X,Y))
    return (im,)

# animation function. This is called sequentially
def animate(i):
    data_slice = z_func(i,X,Y)
    im.set_data(data_slice)
    return (im,)
fig.colorbar(im, shrink=0.5, aspect=5)
#im=ax.imshow(fig, interpolation="spline16", cmap='viridis')
# call the animator. blit=True means only re-draw the parts that have changed.
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=100, interval=40, blit=True)
#rc('animation', html='jshtml')
#anim
HTML(anim.to_html5_video())
```
